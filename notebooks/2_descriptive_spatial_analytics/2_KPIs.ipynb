```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
pd.set_option('display.max_columns', None)
```

```python
def read_all_files(raw=False, sample=False, weather=True):
    if sample:
        if weather:
            data_dir = Path("../../data/input/trip_data_sample.parquet")
            trip_data = pd.read_parquet(data_dir)
            return trip_data
        if not weather:
            data_dir = Path("../../data/weather/trip_data_weather_sample.parquet")
            trip_data = pd.read_parquet(data_dir)
            return trip_data     
    else:
        if raw:
            data_dir = Path("../../data/input/raw/")
            trip_data_df = pd.concat(
                pd.read_parquet(parquet_file)
                for parquet_file in data_dir.glob('*.parquet'))
            trip_data_df.reset_index(inplace=True, drop=True)
            return trip_data_df
        if not raw:
            if weather:
                data_dir = Path("../../data/input/clean/")
                trip_data_df = pd.concat(
                    pd.read_parquet(parquet_file)
                    for parquet_file in data_dir.glob('*.parquet'))
                trip_data_df.reset_index(inplace=True, drop=True)
                return trip_data_df
            if not weather:
                data_dir = Path("../../data/input/weather/")
                trip_data_df = pd.concat(
                    pd.read_parquet(parquet_file)
                    for parquet_file in data_dir.glob('*.parquet'))
                trip_data_df.reset_index(inplace=True, drop=True)
                return trip_data_df

df = read_all_files()
df
```

Trips per Hour of a day 

```python
def plt_clock(ax, radii, title, color):
    """
    Function helping to draw a clock for plotting hourly data
    :param ax: axis for subplots
    :param radii: radius of the clock
    :param title: Title of the plot
    :param color: Color of the drawn data
    :return: Clock plot for plotting hourly data
    """
    n = 24
    bottom = 2

    # create theta for 24 hours
    theta = np.linspace(0.0, 2 * np.pi, n, endpoint=False)

    # width of each bin on the plot
    width = (2 * np.pi) / n

    ax.bar(theta, radii, width=width, bottom=bottom, color=color, edgecolor="#999999")

    # set the lable go clockwise and start from the top
    ax.set_theta_zero_location("N")

    # clockwise
    ax.set_theta_direction(-1)

    # set the label
    ax.set_xticks(theta)
    ticks = ["{}:00".format(x) for x in range(24)]
    ax.set_xticklabels(ticks)
    ax.set_title(title)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(6, 6))
# fig.suptitle('Trips per Start Hour 2013', fontsize=10, fontweight="bold")
axs = plt.subplot( polar=True)

# make the histogram that bined on 24 hour
radii = np.array(df_trips_hour["trip_id"], dtype="int64")
title = "Trips per End Hour Histogram"
plt_clock(axs, radii, title, "#dc143c")

# axs[1] = plt.subplot(2, 2, 2, polar=True)
# radii = np.array(b, dtype="int64")
title = "Trips per End Hour Histogram"

# plt_clock(axs[1], radii, title, "#dc143c")
```
